```
import sys
import base64
from IPython.core.display import display, HTML

class Visualization:

  def __init__(self,path,data):
    """
    @params path for facets/facets_overview/python/ 
    """
    if not isinstance(data,pd.DataFrame):
      raise TypeError('Pass dataframe as argument')
    sys.path.insert(0, path)
    self.data=data
    
    # Create the feature stats for the datasets and stringify it.
    
  def overview_data(self) -> None:
    """
    @params None
    @returns display overview of data.
    """
    from generic_feature_statistics_generator import GenericFeatureStatisticsGenerator
    gfsg = GenericFeatureStatisticsGenerator()
    proto = gfsg.ProtoFromDataFrames([{'name': 'train', 'table': self.data}])
    protostr = base64.b64encode(proto.SerializeToString()).decode("utf-8")

    HTML_TEMPLATE ="""
        <script src="https://cdnjs.cloudflare.com/ajax/libs/webcomponentsjs/1.3.3/webcomponents-lite.js"></script>
        <link rel="import" href="https://raw.githubusercontent.com/PAIR-code/facets/1.0.0/facets-dist/facets-jupyter.html" >
        <facets-overview id="elem"></facets-overview>
        <script>
          document.querySelector("#elem").protoInput = "{protostr}";
        </script>"""
    html = HTML_TEMPLATE.format(protostr=protostr)
    display(HTML(html))
  
  def depthview_data(self) -> None:
    """
    @params None
    @returns display in depth viw of data.
    """
    data=self.data
    from generic_feature_statistics_generator import GenericFeatureStatisticsGenerator
    gfsg = GenericFeatureStatisticsGenerator()
    proto = gfsg.ProtoFromDataFrames([{'name': 'train', 'table': data}])
    protostr = base64.b64encode(proto.SerializeToString()).decode("utf-8")

    jsonstr = data.to_json(orient='records')
    HTML_TEMPLATE = """
            <script src="https://cdnjs.cloudflare.com/ajax/libs/webcomponentsjs/1.3.3/webcomponents-lite.js"></script>
            <link rel="import" href="https://raw.githubusercontent.com/PAIR-code/facets/1.0.0/facets-dist/facets-jupyter.html">
            <facets-dive id="elem" height="600"></facets-dive>
            <script>
              var data = {jsonstr};
              document.querySelector("#elem").data = data;
            </script>"""
    html = HTML_TEMPLATE.format(jsonstr=jsonstr)
    display(HTML(html))


```
